```python
path = "captions_text.vtt"
def read_vtt_file(file_path):
    try:
        with open(file_path) as file:
            content = file.read()
        return content
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return None
read_vtt_file(path)
```

```python
def parse_vtt_content(content):
    time_txt = content.split("\n\n")[1:]
    text = ""
    for ele in time_txt:
        # ele is a string we can split it at \n
        ele_list = ele.strip().split("\n")
        # at index 0 it is serial, at 1 it is time stamp and rest is text
        time_stamp = ele_list[1]
        start_time, end_time = time_stamp.split("-->")
        text += " ".join(ele_list[2:])
    return text
    
parse_vtt_content(content)
```

### Replace all four letter words:
    keep first character and replace other 3 with *

```python
words = final_text.split()
```

```python
clean_text = []
for word in words:
    if len(word) == 3 and word.isalpha():
        word = word[0]+(3*"*")
        clean_text.append(word)
    else:
        clean_text.append(word)
clean_text = " ".join(clean_text)
```

```python
print(clean_text)
```

```python
# Download this clean file as txt
# file name which we are creating is clean_transcript.txt and w is for write
output_file_path = "clean_transcript.txt"
with open(output_file_path, "w") as output_file:
    # write clean_text values over the sepecified file
    output_file.write(clean_text)
    
# clean_transcript.txt file is created    
```
